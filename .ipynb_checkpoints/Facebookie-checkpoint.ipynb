The gist for the project is as such:

Objective: To analyse the kind of likes by people who are engaged with products from certain facebook pages. This will be used to better understand the kind of interests that people have, who are drawn to certain products.

The workflow will be as such:

- Within a Facebook page, look into each post that features a product
  - Within each post, gather the following:
    - Date/Time of post
    - Number of likes/reactions
    - Content of post - Image/Video or just text only, and what product (Text won't actually tell you!)
    - Number of shares
    - Number of comments
    
    - Within each share, if person and within 48 hrs from time of posting, get the following:
      - Content (Do rough sentiment analysis, only analyse neutral/positive)
      - Number of likes/reactions
      - Number of comments
      - Get the following from the person's profile (label as 1st degree):
        - Groups
        - Likes (within a certain category type)
        - Check-ins (places)
      - Number of people tagged
        - Within each person tagged, get the following (label as 2nd degree):
          - Groups
          - Likes (within a certain category type)
          - Check-ins (places)
        - If person commented, label as 1st degree
          
    - Within each comment, if person and within 48hrs from time of posting, get the following:
      - Content (Do rough sentiment analysis, only analyse neutral/positive)
      - Number of likes/reactions
      - Number of replies
      - Get the following from the person's profile (label as 1st degree):
        - Groups
        - Likes (within a certain category type)
        - Check-ins (places)
      - Number of people tagged
        - Within each person tagged, get the following (label as 2nd degree):
          - Groups
          - Likes (within a certain category type)
          - Check-ins (places)
        - If person replied, label as 1st degree

Approach:

1. Create a database with all those post details. Include Page name and post id
2. Create a function to collect a person's details, can compile as dictionary
3. Create a class that collates details from all the shares and comments from each post

I just happened to decide to write this particular function down to check if a particular key exists in a dictionary. Usually when we attempt to call the key by itself, if it does not exist in the particular dictionary then it will throw an error. I am also not in favour of putting in an if/else condition on every call so might as well create a function for it.

```python
def jsoncheck(parent, child, nullvalue=None):
    if child in parent.keys():
        return parent[child]
    else:
        return nullvalue
```

We've managed to gather some of the latest feed data, but for purpose of this project demonstration we shall only use one post for reference. Before we continue, it won't hurt getting a few more metadata, namely number of likes, comments and shares.

```python
def frompost(x, kie):
    r = requests.get(url+x+'/?fields='+kie, params={'access_token': access_token})
    if kie in r.json().keys():
        return r.json()[kie]['count']
    else:
        return 0
df['share_count'] = df['post_id'].apply(lambda x: frompost(x,kie='shares'))
```
